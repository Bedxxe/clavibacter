```python
from IPython.display import Image
from os import chdir
import numpy as np
import gudhi as gd
import matplotlib.pyplot as plt
import argparse
```

Crear un simplex tree basio, para luego ir agregando los simplejos

```python
st = gd.SimplexTree()
```

```python
#agregar 0-simplejos
st.insert([0])
st.insert([1])
st.insert([2])
```

```python
#agregar la arista 0-2 en la filtracion 1
st.insert([0, 2],filtration=1.0)
```

```python
st.insert([0, 1],filtration=2.0)
```

```python
st.insert([1, 2],filtration=3.0)
```

```python
st.insert([0, 1,2],filtration=4.0)
```

```python
st.num_vertices()
```

```python
st.num_simplices()
```

```python
for sk_value in st.get_skeleton(2):
    print(sk_value)
```

```python
st.compute_persistence()
```

```python
st.persistence()
```

```python
diag = st.persistence()
diag = st.persistence(homology_coeff_field=2, min_persistence=0)
print("diag=", diag)

gd.plot_persistence_diagram(diag)

#plt.savefig('persistence_diagram.svg' , dpi=1200)
plt.show()
```

```python
gd.plot_persistence_barcode(diag)
#plt.savefig('persistence_barcode.svg' , dpi=1200)
plt.show()
```
